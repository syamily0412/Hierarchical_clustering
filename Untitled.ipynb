```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import normalize
```

```python
data = pd.read_csv('Wholesale customers data.csv')
```

```python
scaled_data = normalize(data)
```

```python
scaled_data
```

```python
scaled_data.shape
```

```python
new_data = pd.DataFrame(scaled_data, columns = data.columns)
```

```python
new_data.head()
```

```python
import scipy.cluster.hierarchy as shc
```

```python
plt.figure(figsize=(10,7))
plt.title('Dendrogram')
dend = shc.dendrogram(shc.linkage(new_data, method='ward'))# complete(maximum), single(minimum), average
```

```python
plt.figure(figsize=(10,7))
plt.title('Dendrogram')
dend = shc.dendrogram(shc.linkage(new_data, method='ward'))
plt.axhline(y =6, color='r', linestyle='--')
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
model = AgglomerativeClustering(n_clusters=2, linkage='ward')
```

```python
clusters = model.fit_predict(new_data)
```

```python
#clusters
```

```python
plt.figure(figsize=(10,8))
plt.scatter(new_data.Milk, new_data.Grocery, c=clusters)
```

```python
clusters.shape
```

```python
data['clusters'] = clusters
data.head()
```

```python
data_0 = data[new_data.clusters==0]
data_1 = data[new_data.clusters==1]
```
